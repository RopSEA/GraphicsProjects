```python
import pygame
import moderngl
import numpy
import os
import math
```

```python
pygame.init()
os.environ['SDL_WINDOWS_DPI_AWARENESS'] ='permonitorv2'# pygame setup

pygame.display.gl_set_attribute(pygame.GL_CONTEXT_PROFILE_MASK, pygame.GL_CONTEXT_PROFILE_CORE)
pygame.display.gl_set_attribute(pygame.GL_CONTEXT_MAJOR_VERSION,3)
pygame.display.gl_set_attribute(pygame.GL_CONTEXT_MINOR_VERSION,3)
pygame.display.set_mode((800, 600), pygame.OPENGL | pygame.DOUBLEBUF)
pygame.display.set_caption(title = "Assignment 3: Remi Roper")

gl = moderngl.get_context()


def x(t):
    return math.sin(t)*math.cos(t)*math.log(abs(t))
def y(t):
    return math.pow(abs(t),0.3)*math.sqrt(math.cos(t))-0.5

dt = 0.001
t = -1
xy = []

while t < 1.1:
    xy.append([x(t), y(t)])
    t = t + dt


leng = len(xy) -2

positions = []


for x in range(leng):
    positions.append(xy[0])
    positions.append(xy[x+1])
    positions.append(xy[x+2])
 




vertexBuffer = gl.buffer(
    numpy.array(positions).astype("float32")
)


vertex_shader_code= '''
#version 330 core

layout (location = 0) in vec2 position;
uniform float scale;

void main()
{
    gl_Position = vec4(position.x * scale, position.y * scale ,  0.0, 1.0);
}

'''

fragment_shader_code= '''
#version 330 core

out vec4 outColor;

void main()
{
    outColor = vec4(1.0, 0.0, 1.0, 1.0);
}

'''
zoomFactor = 1.0
deltaZoom = 0.025
running = True

program = gl.program(vertex_shader=vertex_shader_code, fragment_shader =fragment_shader_code)

renderable = gl.vertex_array(program, [(vertexBuffer, '2f','position')])
renderable.program["scale"] = zoomFactor                            
gl.clear(0.2,0.2,0.2)
renderable.render()
pygame.display.flip()

while running:
    for event in pygame.event.get(): # poll for events
        if event.type == pygame.QUIT or (event.type == pygame.KEYDOWN and event.key == 27):
            running = False
    gl.clear(0.2, 0.2, 0.2)
    renderable.render()
    renderable.render(mode=moderngl.TRIANGLE_FAN)
    renderable.program["scale"] = zoomFactor

    zoomFactor = zoomFactor + (0.1 * deltaZoom)

    if zoomFactor >= 3:
         deltaZoom = -1 * deltaZoom

    if zoomFactor <= 0:
        deltaZoom = -1 * deltaZoom
    
    pygame.display.flip()

pygame.quit()
```
